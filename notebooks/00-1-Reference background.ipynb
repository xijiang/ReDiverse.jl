## A summary of cattle (*Bos taurus*) assemblies
There are 3 assemblies available for cattle, Btau, UMD and ARS-UCD. ARS-UCD is now available on [NCBI](ftp://ftp.ncbi.nlm.nih.gov/genomes/refseq/vertebrate_mammalian/Bos_taurus/latest_assembly_versions/GCF_002263795.1_ARS-UCD1.2). NCBI also offer [downloads for older assemblies of Btau and UMD](ftp://ftp.ncbi.nlm.nih.gov/genomes/refseq/vertebrate_mammalian/Bos_taurus/all_assembly_versions/suppressed/).

Above links were tracked from [this very brief history of cattle assemblies](https://bovinegenome.elsiklab.missouri.edu/history).
